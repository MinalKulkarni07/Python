### Super Advance Billing System

```python
 #-- User input section
# Enter Product 1 -
# Samosa
# Enter Samosa Price :
# 300/-
# Enter Samosa QT :
# 3
# -----------------------------
# Do you Want to add more [Y/N]
# Y
# Enter Product 2 -
# Kachori
# Enter Kachori Price :
# 100/-
# Enter Kachori QT :
# 2
# ----------------------------
# Do you Want to add more [Y/N]
# Y
# Enter Product 3 -
# Jalebi
# Enter Jalebi Price :
# 200/-
# Enter Jalebi QT :
# 4
# Do you Want to add more [Y/N]
# N

# Do You Want to add GST [Y/N]
# [note : if user select `N` Print bill without GST]
# [Note User select yes `Y`]
# Enter GST % = 18

# Select Offer :
# 1. FLAT 2% OFF
# 2. FLAT 4% OFF
# 3. FLAT 8% OFF
# [Note if user Select offer = then offer apply into bill] 
# if user select [ 1 ]

# -------Out put -------

# 1. Samosa = 900/-
# 2. Kachori = 200/-
# 3. Jalebi = 800/-
# -------------------
# Total = 1900
# GST = 18%
# offer = 2%
# -------------------
# Final Total = 2197.16/-
# -------------------
```

```python
# Super Advance Billing System

data_dict = {}

count = 1 
while True:
    Product = input(f"Enter Product {count}")
    Price = int(input(f"Enter {Product} Price: "))
    Quantity = int(input(f"Enter {Product} Quantity: "))
    data_dict[Product] = {"Price":Price, "Quantity":Quantity}
    
    Add_Product = input("Do you want add more product[Y/N]: ")
    if Add_Product == "Y":
        count += 1
    elif Add_Product == "N":
        break

def total(data_dict):
    return sum(items["Price"] * items["Quantity"]for items in data_dict.values())

def gst(Total, gst_percentage):
    return Total * (gst_percentage / 100)
    
def final_total(Total,GST):
    return (Total + GST)   

def offer(Final_Total,Select_Offer):
    return Final_Total * (Select_Offer/100)

def grand_total(Final_Total,Offer):
    return (Final_Total - Offer)

Total = total(data_dict)
Final_Total = final_total(Total,GST)
Offer = offer(Final_Total,Select_Offer)

Add_GST = input("Do you want to add GST [Y/N]")
GST = 0
if Add_GST == "Y":
    gst_percentage = int(input("Enter GST percentage :"))
    Total = total(data_dict)
    GST = round(gst(Total, gst_percentage),2)
    print(f"GST = {gst_percentage}%")
    print(f"GST = {GST}")

print("""
Offers:
1. FLAT 2% OFF
2. FLAT 4% OFF
3. FLAT 8% OFF""")

Select_Offer = int(input("Select Offer ="))

if Select_Offer == 1:
    Offer = offer(Final_Total,2)
elif Select_Offer == 2:
    Offer = offer(Final_Total,4)
elif Select_Offer == 3:
    Offer = offer(Final_Total,8)
else:
    print("Offer not valid.No offer applied.") 

print("\n------Bill------")
count = 1
for Product, item in data_dict.items():
    print(f"{count}.{Product} = {item['Price']} * {item['Quantity']} = {item['Price'] * item['Quantity']}/-")
    count+= 1
    
print("--------------")
Total = total(data_dict)
print("Total = ",Total)

if Add_GST == "Y":
    print(f"GST = {gst_percentage}%")
else:
    print(f"GST = 0")

if Select_Offer == 1:
    print(f"Offer = 2%")
elif Select_Offer == 2:
    print(f"Offer = 4%")
elif Select_Offer == 3:
    print(f"Offer = 8%")
else:
    print(f"Offer = -") 

print("--------------")
Grand_Total = grand_total(Final_Total,Offer)
print(f"Grand Total = {Grand_Total}/-")
print("--------------")
    
```
