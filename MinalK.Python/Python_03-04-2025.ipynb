```python
# 03 apr 2025
#Line Plot: A simple line plot showing a sine wave.

import matplotlib.pyplot as plt # matplotlib.pyplot is a plotting library — it lets you create visualizations like line charts, bar charts, etc.
import numpy as np # numpy is a numerical computing library — it's used here to generate an array of numbers.

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show()
```

```python
#Scatter Plot: A scatter plot with random data points.
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(60)
y = np.random.rand(60)
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#Bar Plot: A bar plot displaying data with categories.

import matplotlib.pyplot as plt
categories = ['Category A', 'Category B', 'Category C']
values = [30, 20, 35]
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
#Histogram: A histogram showing the distribution of a dataset.

import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)  # Generating random data
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
#Pie Chart: A pie chart displaying the distribution of a dataset.

import matplotlib.pyplot as plt
labels = 'A', 'B', 'C', 'D'
sizes = [22, 35, 45, 12]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures a circular pie
plt.title('Pie Chart')
plt.show()
```

```python
# Box Plot: A box plot for visualizing the distribution of a dataset.

import matplotlib.pyplot as plt
import numpy as np
data = [np.random.normal(0, std, 100) for std in range(1, 5)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

```python
#Heatmap: A heatmap to visualize a matrix of data.

import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(15, 15)
plt.imshow(data, cmap='viridis')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```
